```python
import matplotlib.pyplot as plt

input_file = open("result.txt", "r")

res = input_file.read().strip().replace("[", "")[:-1].split(']\n')

num_thread_options = 4
num_size_options = 15

x_array = [pow(2, i) for i in range(num_size_options)]
for i in range(num_thread_options):
    quick_sort_time = res[i].split(', ')
    brick_sort_time = res[i + num_thread_options].split(', ')
    odd_even_merge_sort_time = res[i + 2 * num_thread_options].split(', ')
    merge_sort_time = res[i + 3 * num_thread_options].split(', ')
    bitonic_sort_time = res[i + 4 * num_thread_options].split(', ')
    radix_sort_time = res[i + 5 * num_thread_options].split(', ')
    
    w0_plot, = plt.plot(x_array, [int(item) for item in quick_sort_time], 'r', label='Quick sort')
    w1_plot, = plt.plot(x_array, [int(item) for item in brick_sort_time], 'b', label='Brick sort')
    w2_plot, = plt.plot(x_array, [int(item) for item in odd_even_merge_sort_time], 'g', label='Odd-even merge sort')
    w3_plot, = plt.plot(x_array, [int(item) for item in merge_sort_time], 'c', label='Merge sort')
    w4_plot, = plt.plot(x_array, [int(item) for item in bitonic_sort_time], 'y', label='Bitonic sort')
    w5_plot, = plt.plot(x_array, [int(item) for item in radix_sort_time], 'm', label='Radix sort')
    
    plt.legend(handles=[w0_plot, w1_plot, w2_plot, w3_plot, w4_plot, w5_plot])
    plt.xlabel('Size of array')
    plt.ylabel('Time for sorting (ms)')
    plt.title('Performance of each parallel algorithm with ' + str(pow(2, i)) + ' thread(s)')
    plt.show()
```
